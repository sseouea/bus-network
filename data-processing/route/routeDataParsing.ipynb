```python
import csv
import math
```

```python
routeArr = []
with open('routeData.csv', encoding='utf-8') as f:
    r = csv.reader(f)
    for line in r:
        if line[0] == 'id':
            continue
        routeArr.append([line[0], line[1], line[2].split('/')])
print(routeArr)
```

```python
stopArr = []
with open('../stop/stopData_parsed.csv', encoding='utf-8') as f:
    r = csv.reader(f)
    for line in r:
        if line[0] == 'id':
            continue
        stopArr.append(line)
print(stopArr)
```

```python
stopIdArr = list(map(lambda stop: stop[0], stopArr))
print(stopIdArr)
```

```python
for route in routeArr:
    route[2] = list(filter(lambda stopId: stopId in stopIdArr, route[2]))
print(routeArr)
```

```python
# remove selfloop
for route in routeArr:
    removeIndexArr = []
    for i in range(len(route[2])-1, 0, -1):
        if route[2][i] == route[2][i-1]:
            removeIndexArr.append(i)
    for removeIndex in removeIndexArr:
        route[2].pop(removeIndex)
```

```python
populationDict = {} 
with open('../population.csv', encoding='utf-8') as f:
    r = csv.reader(f)
    for line in r:
        if line[0] == 'id':
            continue
        populationDict[line[1]] = int(line[2])
print(populationDict)
```

```python
newRouteArr = []
totalScore = 0
for route in routeArr:
    populationScore = 0
    for stopId in route[2]:
        populationScore += populationDict[stopArr[stopIdArr.index(stopId)][4]]
    score = len(route[2]) * populationScore
    newRouteArr.append([route[0], route[1], route[2], score, 0])
    totalScore += score

print(totalScore)
```

```python
totalBusCnt = 155 - len(newRouteArr)
remainedBusCnt = totalBusCnt
for route in newRouteArr:
    busCnt = math.floor(totalBusCnt*route[3]/totalScore) + 1
    route[4] = busCnt
    remainedBusCnt -= busCnt-1
print(remainedBusCnt)
print(newRouteArr)
```

```python
for route in sorted(newRouteArr, key=lambda route: route[3], reverse=True)[:24]:
    route[4] += 1

print(newRouteArr)
```

```python
# check total bus cnt
totalBusCnt = 0
for route in newRouteArr:
    totalBusCnt += route[4]
print(totalBusCnt)
```

```python
parsedRouteArr = []
for i, route in enumerate(newRouteArr):
    parsedRouteArr.append([i, route[1], route[4], '/'.join(map(str, route[2]))])
print(parsedRouteArr)
```

```python
with open('routeData_parsed.csv', '+w', encoding='utf-8', newline='') as stopDataFile:
    stopDataWriter = csv.writer(stopDataFile)
    stopDataWriter.writerow(['id', 'name', 'busCnt', 'stopIdArr'])
    stopDataWriter.writerows(parsedRouteArr)
```
